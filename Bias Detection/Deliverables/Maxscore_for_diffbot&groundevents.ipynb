#### Creating a lookup dictionary for the Event ID and Final Scores

```python
import ast
from collections import defaultdict

# Fix stringified lists
first_df_cropped['article_id'] = first_df_cropped['article_id'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x
)
first_df_cropped['final_scores_MF'] = first_df_cropped['final_scores_MF'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x
)

# Mapping: article_id â†’ list of (score, event_id)
score_map = defaultdict(list)

for _, row in first_df_cropped.iterrows():
    if isinstance(row['article_id'], list) and isinstance(row['final_scores_MF'], list):
        for aid, score in zip(row['article_id'], row['final_scores_MF']):
            score_map[aid].append((score, row['event_id']))

score_map = dict(score_map)
```

#### Group the Event Id with Article IDs

Same event IDs could be repeated across several rows with Unique Article IDs
So like One Event ID, several Article IDs

```python
import ast

def parse_article_id(x):
    if isinstance(x, list):
        return x
    if pd.isna(x):
        return []
    try:
        # Convert string representation of list to actual list
        return ast.literal_eval(x) if isinstance(x, str) else [x]
    except:
        # If parsing fails, just return as list
        return [x]

event_ids_from_first = (
    first_df_expanded.groupby('event_id')['article_id']
    .apply(lambda x: list(pd.unique(x)))
    .reset_index()
)
event_ids_from_first.rename(columns={'event_id': 'event_id_first_ds'}, inplace=True)
```

#### Map the event ID to Actual Events for convenient processing

The lookup map was created to map the Actual Events to the corresponding event IDs for ease of Understanding

```python
# Create a mapping from event_id to Major events
event_to_major = dict(zip(first_df_expanded['event_id'], first_df_expanded['Major events']))

# For each list of event_ids, retrieve corresponding Major events from the mapping
event_ids_from_first['major_events'] = event_ids_from_first['event_id_first_ds'].apply(
    lambda ids: [event_to_major.get(ids, '')] if isinstance(ids, str) else [])

```

{Article ID: [(score1, Event ID1), (score2, Event ID2)]}

#### Filter the scores, event IDs that are greater than 0.4 = threshold
Filter out the scores, event IDs in score map depending on the 0.4 threshold
and spliting the tuples in the score map to individual columns

```python
best_event_for_article = {}

for article_id, score_events in score_map.items():
    # Filter pairs with score > 0.4
    filtered = [(score, eid) for score, eid in score_events if score > 0.4]
    if not filtered:
        best_event_for_article[article_id] = None  # No good match
        continue
    # Pick the max score and its event_id
    best_score, best_event = max(filtered, key=lambda x: x[0])
    best_event_for_article[article_id] = (best_event, best_score)
```

```python
# Convert mapping dict to a DataFrame
mapping_df = pd.DataFrame.from_dict(best_event_for_article, orient='index', columns=['best_event_id', 'best_score'])
```

```python
import ast
import pandas as pd
import numpy as np

def has_valid(x):
    # list/tuple/set â†’ membership
    if isinstance(x, (list, tuple, set)):
        return 'Valid' in x
    # string â†’ try parse list-literal; else compare
    if isinstance(x, str):
        try:
            obj = ast.literal_eval(x)
            if isinstance(obj, (list, tuple, set)):
                return 'Valid' in obj
        except (ValueError, SyntaxError):
            pass
        return x.strip().lower() == 'valid'
    return False

mask_label = final_df['match_Labels'].apply(has_valid)
mask_score = pd.to_numeric(final_df['best_score'], errors='coerce').notna()

final_df['MATCH_LABELS'] = np.where(mask_label & mask_score, 'Valid', 'Invalid')
```

#### We merge the EventIDs and Actual Events 
EventIDs are unique with several article IDs grouped by, along with the Actual Event Titles
Now, we use those Event IDs to merge into the main dataset having unique Article IDs with repeated EventIDs.

```python
final_df = final_df.merge(
    event_ids_from_first[['event_id_first_ds', 'major_events']],
    left_on = 'best_event_id',
    right_on = 'event_id_first_ds',
    how='left'
)

final_df
```
