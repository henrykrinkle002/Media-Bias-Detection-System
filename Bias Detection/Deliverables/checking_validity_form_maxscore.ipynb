```python
first_df_expanded = first_df_cropped.explode('article_id')
len(first_df_expanded)
```

```python
import ast
from collections import defaultdict

# Fix stringified lists
first_df_cropped['article_id'] = first_df_cropped['article_id'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x
)
first_df_cropped['final_scores'] = first_df_cropped['final_scores'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x
)

# Mapping: article_id → list of (score, event_id)
score_map = defaultdict(list)

for _, row in first_df_cropped.iterrows():
    if isinstance(row['article_id'], list) and isinstance(row['final_scores'], list):
        for aid, score in zip(row['article_id'], row['final_scores']):
            score_map[aid].append((score, row['event_id']))

score_map = dict(score_map)
```

```python
first_df_expanded['article_id'] = first_df_expanded['article_id'].explode('article_id')
```

```python
import ast

second_df_cropped['article_id'] = second_df_cropped['article_id'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)
```

```python
import ast

# Step 0: Convert stringified lists to actual lists in first_df_cropped['article_id']
def parse_article_id(x):
    if isinstance(x, list):
        return x
    if pd.isna(x):
        return []
    try:
        # Convert string representation of list to actual list
        return ast.literal_eval(x) if isinstance(x, str) else [x]
    except:
        # If parsing fails, just return as list
        return [x]

event_ids_from_first = (
    first_df_expanded.groupby('event_id')['article_id']
    .apply(lambda x: list(pd.unique(x)))
    .reset_index()
)
event_ids_from_first.rename(columns={'event_id': 'event_id_first_ds'}, inplace=True)
```

```python
event_ids_from_first['event_id_first_ds'][0]
```

```python
# Create a mapping from event_id to Major events
event_to_major = dict(zip(first_df_expanded['event_id'], first_df_expanded['Major events']))

# For each list of event_ids, retrieve corresponding Major events from the mapping
event_ids_from_first['major_events'] = event_ids_from_first['event_id_first_ds'].apply(
    lambda ids: [event_to_major.get(ids, '')] if isinstance(ids, str) else [])

```

```python
score_map
```

```python
best_event_for_article = {}

for article_id, score_events in score_map.items():
    # Filter pairs with score > 0.5
    filtered = [(score, eid) for score, eid in score_events if score > 0.5]
    if not filtered:
        best_event_for_article[article_id] = None  # No good match
        continue
    # Pick the max score and its event_id
    best_score, best_event = max(filtered, key=lambda x: x[0])
    best_event_for_article[article_id] = (best_event, best_score)
```

```python
# Convert mapping dict to a DataFrame
mapping_df = pd.DataFrame.from_dict(best_event_for_article, orient='index', columns=['best_event_id', 'best_score'])
```

```python
mapping_df.index.name = 'article_id'
mapping_df.reset_index(inplace=True)

# Merge with main_df on article_id
final_df = second_df_cropped.merge(mapping_df, on='article_id', how='left')

print(final_df)
```
