```python
# Step 2: Create article_id â†’ event_id mapping for valid matches
valid_articles_map = dict(
    second_df_cropped[second_df_cropped['match_Labels'] == 'Valid'][['article_id', 'event_id']].values
)

# Step 3: Function to get valid event_ids from the article list
import ast

def find_valid_event_ids(article_list_str):
        unique_lists = [int(a) for a in article_list_str if a in valid_articles_map]
        return list(set(unique_lists))

# Step 4: Apply function to valid_events article_ids
valid_events['valid_article_ids_from_matched'] = valid_events['article_id'].apply(find_valid_event_ids)

# Step 5: Add a flag if any valid event_ids were found
valid_events['has_valid_articles'] = valid_events['valid_article_ids_from_matched'].apply(lambda x: len(x) > 0)

print(valid_events)
```

```python
def find_valid_event_ids(article_list_str):
        unique_lists = [valid_articles_map[a] for a in article_list_str if a in valid_articles_map]
        return list(set(unique_lists))

valid_events['valid_event_ids_from_matched'] = valid_events['article_id'].apply(find_valid_event_ids)

valid_events['has_valid_events'] = valid_events['valid_event_ids_from_matched'].apply(lambda x: len(x) > 0)
```
